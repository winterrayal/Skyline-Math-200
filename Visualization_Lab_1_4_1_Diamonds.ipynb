The next graphs are for numerical variables

```python
#Create a histogram of data scientist salary in US dollars.
ds_df['salary_in_usd'].hist()
```

```python
import seaborn as sns
```

```python
#Create a histogram of data scientist salary in US dollars using Seaborn
sns.histplot( x ='salary_in_usd' , data = ds_df)
```

```python
#Create boxplot of data scientist salary in US dollars.
ds_df.boxplot( column = 'salary_in_usd')
```

```python
#Create a boxplot of data scientist salary in US dollars using Seaborn
sns.boxplot( y = 'salary_in_usd' , data = ds_df)
```

```python
#Is there a difference in salary in us dollar among company sizes?  Create of side by side boxplots.
sns.boxplot( x = 'salary_in_usd' , y= 'company_size', data = ds_df)
```

```python
#Is there a difference in salary in us dollar among company sizes?  Create of side by side histograms.
sns.displot(x = 'salary_in_usd', col ='company_size', data =ds_df)
```
